1. Scroll through all forums, and find which one exist, append to list the pages and forum no.
2. for loop through every page
    1. for loop through every title, save it in list
        



```python
import requests
from bs4 import BeautifulSoup 
```

```python
s=[[],[]]
for i in range(20):
    try:
        p=int(BeautifulSoup(requests.get("https://web.archive.org/web/20190417091511/https://www.rooshvforum.com/forum-"+str(i)+"-page-1.html").content).find_all("a",{"class":"pagination_last"})[0].text)
        print(i,p)
        s[0].append(i)
        s[1].append(p)
    except:
        continue
        
```

```python
#with open("traveldata.txt","w") as f:
#    f.write("DATASHEETS \n \n")
    
c=0
for x in s[0]:
    for y in range(1,s[1][c]+1):
        r=requests.get("https://web.archive.org/web/20190417091511/https://www.rooshvforum.com/forum-"+str(x)+"-page-"+str(y)+".html")
        r=BeautifulSoup(r.content)
        try:
            k=r.find_all("a",{"class":"subject_old"})+r.find_all("a",{"class":"subject_new"})
            with open("traveldata.txt","a") as f:
                f.write("Forum "+str(x)+" Page "+str(y)+"\n\n")
            print("pg",x,".",y)
            for o in k:
                with open("traveldata.txt","a") as f:
                    f.write(o.text+"\n")
        except:
            continue
    c+=1
        

```
